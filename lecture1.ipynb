### Using Random Number Generator

```python
from random import randint, seed
from math import sqrt, log

print(f'The Square Root of 4 = {sqrt(4)}')
print(f'The Log of 4 = {log(4)}')

seed() # Seed the Random Number Generator
randNum = randint(1, 10) 
print(f'randNum = {randNum}')

seed(0) # Use a seed when you need a random number, but the number must be the same for each instance
randNum1 = randint(1, 10)

seed(1) # using same seed number will result in the same result for each variable
randNum2 = randint(1, 10)

print(randNum1, randNum2)
```

```python
# Print all numbers between 1 and 100 that cannot be divided by 2 or 3, 
# do not end in a 7 or 9, and do not consist of two equal digits

for num in range(1, 101):
    if num % 2 == 0: # if number is even
        continue
    if num % 3 == 0: # if number is divisible by 3
        continue
    if num % 10 == 7: # if number ends in a 7
        continue
    if num % 10 == 9: # if number ends in a 9
        continue
    if num > 9:
        if num % 10 == int(num / 10): # excludes 11, 55, ...
            continue
    print(num)
```

```python
from sys import exit
print("Enter 2 integers between 0 and 1000")
while True:
    x = int(input("Enter Number 1: "))
    if x == 0:
        break
    y = int(input("Enter Number 2: "))
    if y == 0:
        break
    if(x < 0 or x > 1000):
        print("The numbers need to be between 0 and 1000")
        continue
    if(x % y == 0 or y % x == 0):
        print("Error: The Numbers Cannot Be Dividers")
        exit()
    print(x, "*", y, "=", x * y)
print("Goodbye!!")
```

### Rolling Dice with randint()

```python
from random import randint

tests = 1000000
success = 0

for testi in range(tests):
    for d in range(5):
        if(randint(1, 6)) != 6:
            break
        else: # loop did not encounter a break statement, so no 6 was rolled
            success += 1

print("The chance of rolling Five 6's =", success / tests)
print("The theoretical answer =", 1 / (6 ** 5))
```

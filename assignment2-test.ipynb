```python
# Import Libraries:

from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
from random import randint

%matplotlib inline
```

## Function to Create Subplots:

```python
def img_plt(images, labels):
    plt.figure(figsize=(12, 5.5))
    for i in range (1, 11):
        plt.subplot(2, 5, i)
        plt.imshow(images[i - 1, :, :], cmap = 'gray')
        #plt.suptitle('Sample Title') Need to Add a Title
        plt.title('Label: ' + str(labels[i - 1]))
    plt.show()
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Select 10 images from the training set
print('Selecting 10 Images From Training Set...')
x_train_rnd = np.zeros((10, 28, 28))
y_train_rnd = np.arange(10)

for digit in range(10):
    x_train_d = x_train[y_train == digit,:,:]
    #x_train_rnd[digit,:,:] = x_train_d[0,:,:] # Select First Digit From Set
    x_train_rnd[digit,:,:] = x_train_d[randint(0, x_train_d.shape[0]),:,:] # Select Random Digit From Set
img_plt(x_train_rnd, y_train_rnd)
```

```python
# Select 10 images from the testing set
print('Selecting 10 Images From Testing Set...')
x_test_rnd = np.zeros((10, 28, 28))
y_test_rnd = np.arange(10)

for digit in range(10):
    x_test_d = x_test[y_test == digit,:,:]
    #x_test_rnd[digit,:,:] = x_test_d[0,:,:] # Select First Digit From Set
    x_test_rnd[digit,:,:] = x_test_d[randint(0, x_test_d.shape[0]),:,:] # Select Random Digit From Set
img_plt(x_test_rnd, y_test_rnd)
```

```python
x_train_01 = x_train[np.logical_or(y_train == 0, y_train == 1),:,:] # Select 1's and 0's from the training set.
y_train_01 = y_train[np.logical_or(y_train == 0, y_train == 1)]

img_plt(x_train_01[0:10,:,:], y_train_01[0:10]) # Select First 10 images from the set
img_plt(x_train_01[0:10, 12:16, 12:16], y_train_01[0:10])

print(np.sum(x_train_01[0, 12:16, 12:16])) # Sum up the pixels in the image
```
